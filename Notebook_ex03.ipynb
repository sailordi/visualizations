```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_matplotlib = pd.DataFrame({
    'x_values': [0, 4, 5, 6, 7],
    'y_values': [2, 6, 3, 6, 3]
})
```

```python
# Create the figure and axis objects
fig, ax = plt.subplots()

# Plotting the points as blue circles with size of 12 using data from DataFrame
ax.scatter(df_matplotlib['x_values'], df_matplotlib['y_values'], c='blue', s=12, label='Points')

# Plotting the points with red dashdot line and width of 3 using data from DataFrame
ax.plot(df_matplotlib['x_values'], df_matplotlib['y_values'], linestyle='--', color='red', linewidth=3)

# Setting the title and labels
ax.set_title('Display Maker')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')

# Limiting the axes
ax.set_xlim([1, 8])
ax.set_ylim([1, 8])

# Showing ticks on all sides
ax.tick_params(axis='both', which='both', top=True, right=True, labeltop=False, labelright=False)

print("Question 1")
# Show the plot
plt.show()
```
