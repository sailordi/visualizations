```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data for the plot
left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]
```

```python
# Create the figure and the first axis
fig, ax1 = plt.subplots()

# Plotting left data in black on the first axis
ax1.plot(x_axis, left_data, 'k-', label='Left Data', linewidth=2)
# ax1.set_xlabel('X-axis')
ax1.set_ylabel('Big', color='k')
ax1.tick_params(axis='y', labelcolor='k')

# Creating the second axis sharing the same x-axis
ax2 = ax1.twinx()

# Plotting right data in red on the second axis
ax2.plot(x_axis, right_data, 'r-', label='Right Data', linewidth=2)
ax2.set_ylabel('Small')
ax2.tick_params(axis='y')

# Adding the title
plt.title('Twin axes')

print("Question 1")
# Show the plot
plt.show()
```
