```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

# Adjust spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Loop to populate each subplot
for i in range(2):
    for j in range(3):
        index = i * 3 + j + 1  # Calculate the index for titles and texts
        axs[i, j].text(0.5, 0.5, f"(2,3,{index})", fontsize=12, ha='center')
        axs[i, j].set_title(f"Title {index}")

print("Question 1")
# Show the plot
plt.show()
```
