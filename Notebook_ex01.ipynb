```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'name':['christopher','marion','maria','mia','clement','randy','remi'],
    'age':[70,30,22,19,45,33,20],
    'gender':['M','F','F','F','M','M','M'],
    'state':['california','dc','california','dc','california','new york','porto'],
    'num_children':[2,0,0,3,8,1,4],
    'num_pets':[5,1,0,5,2,2,3]
    })
```

```python
# Plotting the bar chart with names on x-axis and age on y-axis, adding a legend and customizing colors
plt.figure(figsize=(10, 6))
bars = plt.bar(df['name'], df['age'])
plt.xlabel('name')
plt.title('Age per name')

# Adding a legend
plt.legend([bars], ['Age'], loc='upper right')

print("Question 1")
# Displaying the plot
plt.show()
```
